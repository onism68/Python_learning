```python
import requests
from bs4 import BeautifulSoup

url = 'https://search.jd.com/Search?keyword=机械键盘&enc=utf-8'
res = requests.get(url)
res.encoding = 'utf-8'
te = res.text
goodsList = BeautifulSoup(te,'html.parser').select('.clearfix')[0]('ul')[0]('li')
for g in range(len(goodsList)):
    pNameType2 = goodsList[g]('a')[1].text.strip()
    pPrice = goodsList[g]('div')[3].text.strip()
    pCommit = goodsList[g].select('.p-commit')[0]('strong')[0]('a')[0].text
    J_im_icon = goodsList[g].select('.curr-shop')[0].text
    print(pNameType2,pPrice,J_im_icon,pCommit)
```
