```python
import json
import requests



# 获取电影评论
def getComments(movieId):
    hotCommeUrl1 = 'https://api-m.mtime.cn/Showtime/HotMovieComments.api?pageIndex=1&movieId={}'
    hotCommeUrl = hotCommeUrl1.format(movieId)
    commentsRes = requests.get(hotCommeUrl)
    commentsJson = json.loads(commentsRes.text)
    commentsData = commentsJson['data']['cts']
    for c in range(len(commentsData)):
        ca = commentsData[c]['ca']
        ce = commentsData[c]['ce']
        cd = commentsData[c]['cd']
        print(movieId,ca,cd,ce)

        
# 获取电影列表
def getMovieList():
    listUrl = 'https://api-m.mtime.cn/PageSubArea/HotPlayMovies.api?locationId=791'
    listRes = requests.get(listUrl)
    movieJson = json.loads(listRes.text)
    movieList = movieJson['movies']
    for m in range(len(movieList)):
        movieId = movieList[m]['movieId']
        getComments(movieId)
#         titleCn = movieList[m]['titleCn']
#         movieType = movieList[m]['actorName1']
#         actorName1 = movieList[m]['type']
#         actorName2 = movieList[m]['actorName2']
#         directorName = movieList[m]['directorName']
#         print(movieId,titleCn,movieType,directorName,actorName1,actorName2)
```
