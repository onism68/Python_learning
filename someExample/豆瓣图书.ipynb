```python
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient
import time
import numpy as np

conn= MongoClient(host="127.0.0.1",port=27017)
# 代理IP数据库
db1 = conn.ip
my_set1 = db1.ip001
# 书本数据数据库
db = conn.books
my_set = db.books001

# 获取图书标签
def getTags():
    tagurl = 'https://book.douban.com/tag/?view=cloud'
    res = requests.get(tagurl)
    bs = BeautifulSoup(res.text,'html.parser')
    article = bs.select('.article')[0].select('.tag_cloud')
    tbody = article[0]('td')
    for t in range(len(tbody)):
        tag = tbody[t]('a')[0].text
        getBooks(tag)
        time.sleep(5)

#获取每本书的相关信息

def getBooks(tag):
    proxie=[]    
    # 生成随机数，用于随机取出数据库中的代理IP
    a = np.random.randint(1,35)
    for i1 in my_set1.find()[a-1:a]:
        ip = i1['ip']
        port = i1['port']
        print(ip)
        proxie.append({
            'http':ip+':'+port,
        })
    proxies = proxie[0]
    print(proxies)
    bookurl0 = 'https://book.douban.com/tag/{}'
    bookurl = bookurl0.format(tag)
    bres = requests.get(bookurl,proxies = proxies)
    bbs = BeautifulSoup(bres.text,'html.parser')
    blist = bbs.select('.subject-list')[0]('li')
    for b in range(len(blist)):
        subJect_id = blist[b].select('.info')[0]('h2')[0]('a')[0]['href'].lstrip('https://book.douban.com/subject/').rstrip('/')
        bTitle = blist[[b].select('.info')[0]('h2')[0]('a')[0]['title']
        littleTitle = blist[[b].select('.info')[0]('h2')[0]('a')[0]('span')[0].text.strip()
        title = str(bTitle)+str(littleTitle)
        pub = blist[b].select('.info')[0].select('.pub')[0].text.strip()
        con = blist[b].select('.info')[0]('p')[0].text.strip()
        star = blist[b].select('.info')[0].select('.clearfix')[0]('span')[0]['class'][0].lstrip('allstar')
        rating_nums = blist[b].select('.info')[0].select('.clearfix')[0]('span')[1].text
        p1 = blist[b].select('.info')[0].select('.clearfix')[0]('span')[2].text.strip()
        my_set.insert({"b":b,"subJect_id":subJect_id,"title":title,"con":con,"star":star,"rating_nums":rating_nums,"p1":p1})
        print(b,subJect_id,title,pub,con,star,rating_nums,p1)
```
