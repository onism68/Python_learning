```python
import json
import time
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient
##数据库连接
conn= MongoClient(host="127.0.0.1",port=27017)
db = conn.books
my_set = db.books001

def getBooks(s):
    urlb = 'https://book.douban.com/top250?start={}'
    url = urlb.format(s)
    res = requests.get(url)
    bs = BeautifulSoup(res.text,'html.parser')
    article = bs.select('.indent')[0]('td')
    for i in range(1,len(article),2):
        bid = article[i]('a')[0]['href'].lstrip('https://book.douban.com/subject/').rstrip('/')
        title = article[i]('a')[0]['title']
        pl = article[i].select('.pl')[0].text
        star = int(article[i].select('.clearfix')[0]('span')[0]['class'][0].lstrip('allstar'))/10
        rating_nums = article[i].select('.rating_nums')[0].text
        pj = article[i].select('.clearfix')[0].select('.pl')[0].text.lstrip('(').rstrip(')').strip().rstrip('人评价')
        #inq = article[i].select('.inq')[0].text
        #入库操作
        my_set.insert({"书本id":bid,"名称":title,"相关信息":pl,"星级":star,"评分":rating_nums,"评价数":pj})
        print(bid,title,pl,star,rating_nums,pj)

for s in range(0,250,25):
    getBooks(s)
    time.sleep(5)
```

```python
title = []
rating_nums = []
for i in my_set.find():
    title1 = i['名称']
    rating_num = i['评分']
    title.append(title1)
    rating_nums.append(rating_num)
print(title)
```

```python
from pyecharts import Line

line = Line("×××","测试")
#is_label_show是设置上方数据是否显示
line.add("评分", title, rating_nums, is_label_show=True)
#line.add("蒸发量", columns, data2, is_label_show=True)
line.render('a.html')
```
